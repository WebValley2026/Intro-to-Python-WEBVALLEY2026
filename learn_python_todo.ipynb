## 11. Opening radar data

Real weather and **radar** data is usually **datacube**: a value stored for many points on a map (latitude × longitude) *and* for many moments in time.

```
          time ---->
   lat  +-----------+
    |   |           |   each square = a measurement
    v   |   grid    |   at one place and one time
        +-----------+
             lon
```

**xarray** handles these cubes with *named* dimensions (`time`, `lat`, `lon`) and coordinates, so the data stays easy to understand — it's NumPy + labels.

### How you open a real radar file

Radar/weather files are usually **NetCDF** (`.nc`) or GRIB. Weather experts are transitioning to Zarr format, but that's a topic for Giacomo, also called "The Zarrifier". With xarray by the way it's one line:

### 🍄 The final boss: real data!

Mario's 3×3 toy map was cute, but real radar is *much* bigger. Bowser scattered storms all over **Italy**, and Mario needs to track the rain.

The radar gives him **rain rate** (how much rain falls per hour — higher = heavier rain) for a **1400 × 1200** grid of Italy, at **over a million moments in time** (from 2010 to 2025, every 10–15 minutes).

That's a datacube: `time × y × x`. The `y` and `x` dimensions are projected coordinates in metres. `lat` and `lon` also exist as 2D auxiliary coordinates, but the *indexable* axes are `y` and `x`. NumPy alone would force Mario to *remember* which axis is which. **xarray keeps the names**, so he never gets lost.

````python
import xarray as xr

# Mario gets his hands on REAL radar data: a scan over Trentino.
ds = xr.open_dataset("/data/radar/italian-radar-dpc-sri.zarr")  # lazy: only metadata read, numbers still on disk
ds                                      # show the Dataset: dims, coords and variables
````

### Wait — why was opening so fast? (xarray is *lazy*)

When you call `xr.open_dataset(...)`, xarray does **not** read the numbers into memory. It only reads the *metadata*: the variable names, the dimensions, the coordinates and the units. The actual data stays on disk until you really need it.

This is called **lazy loading**. It's why opening a 50 GB file is instant.

```
   xr.open_dataset("radar_data.nc")
              │
              │  reads metadata only (dims, coords, units)
              ▼
        Dataset object  ─ ─ ─ ─▶  numbers still on disk
              │
              │  data read into RAM only when you ask for a result:
              ▼
     .values    .plot()    .mean()    .compute()
```

**Why it's great:**
- Open huge files "instantly" (metadata is small).
- Slice/select first (`.isel`, `.sel`), so you load *only the bytes you need*.
- Chain operations cheaply; nothing runs until you ask for a result.

This is really important when you work with **big data**: you can open a 50 GB file, select a small region and time range, and compute the average without ever loading the whole file into memory. Weather data can easily reach hundreds of GBs or even TBs!!! And RAM these days its quite expensive...


### Working with the cube: select and reduce

````python
# Grab ONE variable out of the dataset. It keeps its dims and coords with it.
rainrate = ds["RR"]
print("Dimensions:", rainrate.dims)    # ('time', 'y', 'x')
print("Shape:", rainrate.shape)        #
rainrate                               # a labelled NumPy array = a DataArray
````

````python
first_scan = rainrate.isel()              # position 0 → first time step
print(first_scan.shape)  

point_isel = rainrate.isel(y=, x=)       
print(point_isel.values)

# --- .sel: select by LABEL ---
first_time = rainrate.time.values[0]           # e.g. 2010-01-01T00:10:00
print(first_time)

scan_by_time = rainrate.sel(time=)   # same scan as isel
print(scan_by_time.shape)  # (1400, 1200)
````
