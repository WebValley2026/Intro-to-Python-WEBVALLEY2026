# Part 1 — Basics

## 1. Types

Every value in Python has a **type**. The basic ones:

| Type | What it is | Example |
|------|------------|---------|
| `int` | whole number | `2026` |
| `float` | number with decimals | `21.5` |
| `str` | text (a "string") | `"Trento"` |
| `bool` | yes/no value | `True` / `False` |

Use `type(value)` to ask Python what type something is.

**Why types matter** — the type decides what you can do with a value. `3 + 2` gives `5` (math), but `"3" + "2"` gives `"32"` (text glued together), and `"3" + 2` is an **error**. Same symbol, different result. Pick the wrong type → wrong answer or a crash.

## 2. Variables

A **variable** is a name that stores a value. Think of it as a labelled box.

**Why it matters** — name a value once and reuse it everywhere; change it in one place and everything updates. Without variables you'd retype the same number all over and edits would be error-prone.

## 3. Operations

You can **compute** with values. Numbers use math operators like `+`, `-`, `*`, `/`.

**Why it matters** — operations turn stored values into answers. The *same* operator acts differently per type (`+` adds numbers but glues text), so the type tells you what will happen.

Strings have their own tricks.

# Part 2 — Data structures

## 4. Lists

A **list** stores many values in a single variable, kept in the order they were added. Use square brackets `[ ]`.

**Why it matters** — real data is many values (every temperature, every city). A list keeps them in one variable so you can loop over them, instead of inventing a separate name for each.

## 5. Dictionaries

A **dictionary** stores values with a *label* (called a **key**) instead of a number position. Use curly braces `{ }` with `key: value` pairs.

**Why it matters** — labels are clearer than positions: `weather["temp"]` says what it is, `data[3]` doesn't. Ideal for one record with named fields.

# Part 3 — Control flow

## 6. if / else, for loops, while loops

**Control flow** lets your program make decisions and repeat work.

- `if` / `elif` / `else` — do different things depending on a condition
- `for` — repeat once for every item in a list (or range of numbers)
- `while` — keep repeating *as long as* a condition stays `True`

Important: in Python the **indentation** (the spaces at the start of a line) decides what is *inside* the block.

# Part 4 — NumPy and Pandas

## 7. NumPy arrays

**NumPy** is a library for fast math on many numbers at once. Its main tool is the **array**, similar to a list but much faster and able to do math on every value in one line.

We import it with the nickname `np` (everybody does this).

### 2D arrays (a grid of numbers)

An array can have rows *and* columns, like a table or a map. This is the idea behind a weather map: a value for every point of a grid.

## 8. Pandas: tabular data

**Pandas** handles **tables** (rows and columns), like a spreadsheet in code. The main object is the **DataFrame**.

It's perfect for data with labels and mixed types: one column of cities, one of temperatures, one of dates…

Import it with the nickname `pd`.

# Part 5 — Functions, classes and objects

## 9. Functions

A **function** is a reusable block of code with a name. You write it once, then *call* it whenever you need it — no copy-pasting.

Define one with `def`. Values you pass in are called **arguments**; the value it gives back uses `return`.

# Part 6 — Xarray

## 11. Opening radar data

Real weather and **radar** data is a **datacube**: a value stored for many points on a map (latitude × longitude) *and* for many moments in time.

```
          time ---->
   lat  +-----------+
    |   |           |   each square = a measurement
    v   |   grid    |   at one place and one time
        +-----------+
             lon
```

**xarray** handles these cubes with *named* dimensions (`time`, `lat`, `lon`) and coordinates, so the data stays easy to understand — it's NumPy + labels.

### How you open a real radar file

Radar/weather files are usually **NetCDF** (`.nc`) or GRIB. Weather experts are transitioning finally to Zarr format (and we are going to use it!), but that's a topic for Giacomo, also called "The Zarrifier". With xarray by the way it's one line:

````python
ds = xr.open_dataset("radar_data.nc")  # open the file (lazy: nothing loaded yet)
````

### Wait — why was opening so fast? (xarray is *lazy*)

When you call `xr.open_dataset(...)`, xarray does **not** read the numbers into memory. It only reads the *metadata*: the variable names, the dimensions, the coordinates and the units. The actual data stays on disk until you really need it.

This is called **lazy loading**. It's why opening a 50 GB file is instant.

```
   xr.open_dataset("radar_data.nc")
              │
              │  reads metadata only (dims, coords, units)
              ▼
        Dataset object  ─ ─ ─ ─▶  numbers still on disk
              │
              │  data read into RAM only when you ask for a result:
              ▼
     .values    .plot()    .mean()    .compute()
```

**Why it's great:**
- Open huge files "instantly" (metadata is small).
- Slice/select first (`.isel`, `.sel`), so you load *only the bytes you need*.
- Chain operations cheaply; nothing runs until you ask for a result.

This is really important when you work with **big data**: you can open a 50 GB file, select a small region and time range, and compute the average without ever loading the whole file into memory. Weather data can easily reach hundreds of GBs or even TBs!!! And RAM these days its quite expensive...
