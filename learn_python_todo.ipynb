# Part 1 — The basics, with Mario and friends!

Types, variables, operations, lists and dictionaries — all through one story.
Read each cell, **predict the output**, then run it.

Mario has 10 euros. He buys a sandwich for 3.5 euros and a drink for 1.25 euros (they were cheaper back then!).

````python
# VARIABLES: store each value in a labelled box so we can reuse it.
wallet   = 10      # int   -> a whole number
sandwich = 3.5     # float -> a number with decimals
drink    = 1.25    # float

# TYPES: ask Python what each value is.
print(type(wallet))    # <class 'int'>
print(type(sandwich))  # <class 'float'>

# OPERATION (subtraction): turn the values into an answer.
# Mixing an int with floats gives a float back.
money_left = wallet - sandwich - drink
print("Mario has", money_left, "euros left")  # 5.25
print(type(money_left))                        # <class 'float'>
````

Luigi comes along and gives Mario a banana for free, but is asking him to spell his name backwards.

````python
# STRING (str): text goes in quotes.
name = "Mario"
print(type(name))   # <class 'str'>

# STRING TRICK (slicing): [::-1] walks the text one step at a time, backwards.
backwards = name[::-1]
print(backwards)    # "oiraM"
````

Mario wants to know how much money Luigi owes the bar (Luigi is famously forgetful).

Luigi: "I owe the bar half of what you spent, but I can't remember how much that was."

````python
# OPERATION (addition): how much did Mario spend in total?
spent = sandwich + drink     # 3.5 + 1.25 = 4.75
print("Mario spent", spent, "euros")

# OPERATION (division): Luigi owes half of that.
owed = spent / 2             # 4.75 / 2 = 2.375
print("Luigi owes the bar", owed, "euros")
````

Peach comes along and tells Luigi that his name is wrong! The real name is "gigi", and she offered to change it!

````python
# VARIABLES can be reassigned: a box can hold a new value at any time.
new_name = "gigi"
print("Old name:", name, "-> New name:", new_name)

# STRING OPERATION (+ glues text together, it does NOT do math on text).
print("Peach changed " + name + " into " + new_name)
````

Gigi now is happy, but it is now vital for him to know the difference in length between his old name and his new name.


````python
# STRING TRICK (len): counts how many characters are in the text.
old_len = len(name)        # "Mario" -> 5
new_len = len(new_name)    # "gigi"  -> 4
print("Old name length:", old_len)
print("New name length:", new_len)

# OPERATION (subtraction) on those counts.
difference = old_len - new_len
print("Length difference:", difference)   # 1
````

Peach is now asking money to gigi for the name change and she wants double what he owes the bar.

Gigi is a bit confused, but he does the math anyway.


````python
# OPERATION (multiplication): double = times 2.
price = owed * 2     # 2.375 * 2 = 4.75
print("Gigi must pay Peach", price, "euros")
````

Just a more formal definition...

A **class** is a blueprint for making your own kind of object. An **object** bundles together **data** (variables) and **actions** (functions) that belong to it.

Example: a `WeatherStation` that stores a city name and its temperature readings, and knows how to report them.

- A function that lives inside a class is called a **method**.
- `__init__` is a special method that runs when you create a new object.
- `self` means *this particular object*.

## 11. Opening radar data

Real weather and **radar** data is a **datacube**: a value stored for many points on a map (latitude × longitude) *and* for many moments in time.

```
          time ---->
   lat  +-----------+
    |   |           |   each square = a measurement
    v   |   grid    |   at one place and one time
        +-----------+
             lon
```

**xarray** handles these cubes with *named* dimensions (`time`, `lat`, `lon`) and coordinates, so the data stays easy to understand — it's NumPy + labels.

### How you open a real radar file

Radar/weather files are usually **NetCDF** (`.nc`) or GRIB. Weather experts are transitioning to Zarr format, but that's a topic for Giacomo, also called "The Zarrifier". With xarray by the way it's one line:

### 🍄 The final boss: real data!

Mario's 3×3 toy map was cute, but real radar is *much* bigger. Bowser scattered storms all over **Italy**, and Mario needs to track the rain.

The radar gives him **rain rate** (how much rain falls per hour — higher = heavier rain) for a **1400 × 1200** grid of Italy, at **over a million moments in time** (from 2010 to 2025, every 10–15 minutes).

That's a datacube: `time × y × x`. The `y` and `x` dimensions are projected coordinates in metres. `lat` and `lon` also exist as 2D auxiliary coordinates, but the *indexable* axes are `y` and `x`. NumPy alone would force Mario to *remember* which axis is which. **xarray keeps the names**, so he never gets lost.

````python
import xarray as xr

# Mario gets his hands on REAL radar data: a scan over Trentino.
ds = xr.open_dataset("/data/radar/italian-radar-dpc-sri.zarr")  # lazy: only metadata read, numbers still on disk
ds                                      # show the Dataset: dims, coords and variables
````

### Wait — why was opening so fast? (xarray is *lazy*)

When you call `xr.open_dataset(...)`, xarray does **not** read the numbers into memory. It only reads the *metadata*: the variable names, the dimensions, the coordinates and the units. The actual data stays on disk until you really need it.

This is called **lazy loading**. It's why opening a 50 GB file is instant.

```
   xr.open_dataset("radar_data.nc")
              │
              │  reads metadata only (dims, coords, units)
              ▼
        Dataset object  ─ ─ ─ ─▶  numbers still on disk
              │
              │  data read into RAM only when you ask for a result:
              ▼
     .values    .plot()    .mean()    .compute()
```

**Why it's great:**
- Open huge files "instantly" (metadata is small).
- Slice/select first (`.isel`, `.sel`), so you load *only the bytes you need*.
- Chain operations cheaply; nothing runs until you ask for a result.

This is really important when you work with **big data**: you can open a 50 GB file, select a small region and time range, and compute the average without ever loading the whole file into memory. Weather data can easily reach hundreds of GBs or even TBs!!! And RAM these days its quite expensive...


### Working with the cube: select and reduce

Mario doesn't always want the whole cube. xarray lets him **select** a piece (one time, one place) and **reduce** it (average, max) using the *dimension names* — no guessing which axis is which.

````python
# Grab ONE variable out of the dataset. It keeps its dims and coords with it.
refl = ds["RR"]
print("Dimensions:", refl.dims)    # ('time', 'y', 'x')
print("Shape:", refl.shape)        #
refl                               # a labelled NumPy array = a DataArray
````

````python
# ── WHEN TO USE EACH ─────────────────────────────────────────────────────────
#  .isel(dim=N)          → you know the POSITION (like list[0])
#  .sel(dim=value)       → you know the LABEL   (the real coordinate value)
#
#  Rule of thumb:
#    "I want the 1st scan"           → isel(time=0)
#    "I want the scan from Jan 2010" → sel(time="2010-01-01T00:00")
#    "I want grid row 700, col 600"  → isel(y=700, x=600)
#    "I want the point at y=4.8e6 m" → sel(y=4_800_000, x=300_000, method="nearest")
# ─────────────────────────────────────────────────────────────────────────────

# --- .isel: select by INTEGER POSITION ---
first_scan = refl.isel(time=0)              # position 0 → first time step
print("isel | first scan shape:", first_scan.shape)  # (1400, 1200)

point_isel = refl.isel(y=700, x=600)       # row 700, column 600 — no idea what coords those are
print("isel | rain at grid[700,600], first 5 steps:", point_isel.values[:5].round(2))

# --- .sel: select by LABEL (actual coordinate value) ---
# Peek at what timestamps and coordinates actually exist:
first_time = refl.time.values[0]           # e.g. 2010-01-01T00:10:00
print("\nFirst timestamp in the data:", first_time)

scan_by_time = refl.sel(time=first_time)   # same scan as isel(time=0), but readable
print("sel  | same scan shape:", scan_by_time.shape)  # (1400, 1200)

# For y/x we need method="nearest" because we rarely hit an exact grid value:
y_coord = float(refl.y.values[700])        # grab the real metre-value at row 700
x_coord = float(refl.x.values[600])        # grab the real metre-value at col 600
point_sel = refl.sel(y=y_coord, x=x_coord, method="nearest")
print("sel  | rain at y={:.0f} m, x={:.0f} m, first 5 steps:".format(y_coord, x_coord),
      point_sel.values[:5].round(2))

# Both give the same numbers — the difference is only HOW you express the index:
print("\nSame result?", (point_isel.values == point_sel.values).all())
````

````python
# REDUCE over a NAMED dimension -- no need to remember "axis 0" vs "axis 2".
avg_map = refl.mean(dim="time")             # average rain rate map over all time steps
print("Average map shape:", avg_map.shape)  # (1400, 1200) -> time is gone, a 2D map is left

# Strongest rain anywhere on the map, for each moment in time:
peak = refl.max(dim=["y", "x"])
print("Peak rain rate (first 5 steps):", peak.values[:5].round(2))
````

## 12. Wonderful plots

The best part of xarray: because the data knows its own dimensions and coordinates, **plotting is one line** and the axes, labels and colorbars come out right automatically.

````python
import matplotlib.pyplot as plt

# ONE LINE. Because the data carries its own lat/lon coords, xarray labels the
# axes and adds a colorbar automatically -> a real radar map of Trentino.
refl.isel(time=0).plot()
plt.title("Radar reflectivity over Trentino — first scan")
plt.show()
````

````python
# A 2D result -> a map. The time-averaged rain rate, same one-liner.
refl.mean(dim="time").plot(cmap="viridis")
plt.title("Average rain rate (all scans)")
plt.show()

# A 1D result -> xarray draws a line automatically: peak storm strength over time.
refl.max(dim=["y", "x"]).plot()
plt.title("Peak rain rate over time")
plt.show()
````
